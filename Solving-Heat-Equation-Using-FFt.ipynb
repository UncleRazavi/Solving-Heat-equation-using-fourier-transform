```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import axes3d
```

```python
a = 1 
L = 100 
N=1000
dx = L/N
x = np.arange(-L/2 , L/2 , dx) 
```

```python
omega = 2 * np.pi*np.fft.fftfreq(N , d=dx) #np.fft.fftfreq(N , d = timespecs)
#Return the Discrete Fourier Transform sample frequencies.
```

```python
omega
```

```python
#initial condition 
u0 = np.zeros_like(x)
```

```python
u0
```

```python
#to create a pulse signal
u0[int((L/2 - L/10)/dx): int((L/2+L/10)/dx)]=1
```

```python
u0
```

```python
u0hat = np.fft.fft(u0)
```

```python
u0hat_ri = np.concatenate((u0hat.real,u0hat.imag))
```

```python
dt = 0.1
t = np.linspace(0,10,int(10/dt)+1)

def rhsHeat(u0hat_ri, t, omega, a):
    uhat = u0hat_ri[:N] + 1j * u0hat_ri[N:]
    d_uhat = -a**2 * (np.power(omega, 2)) * uhat
    duhat_ri = np.concatenate((d_uhat.real, d_uhat.imag))
    return duhat_ri

uhat_ri = odeint(rhsHeat, u0hat_ri, t, args=(omega, a))

uhat_ri = uhat_ri[:, :N] + (1j) * uhat_ri[:, N:]

u = np.zeros_like(uhat_ri)

for k in range(len(uhat_ri)):
    u[k, :] = np.fft.ifft(uhat_ri[k, :])

u = u.real

print(u)

```

```python

time_indices = np.linspace(0, len(t)-1, 4, dtype=int)  # Choosing 4 equally spaced time indices

# Create waterfall plot using matplotlib
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each selected time step
for i in time_indices:
    ax.plot(np.arange(N), [t[i]] * N, u[i, :], color='blue')

ax.set_xlabel('Spatial Grid Points')
ax.set_ylabel('Time')
ax.set_zlabel('u')
ax.set_title('Waterfall Plot of u (4 time points)')

plt.show()
```
